```python
myfile=open("/Users/yang/Desktop/adventofcode/heightmap.txt",'r')
lines=myfile.readlines()

S_letter="S"
E_letter="E"

#find S and E location
#replace S and E to a and z
for i in range(len(lines)):
    lines[i]=lines[i].strip()
    countS = lines[i].count(S_letter)
    countE = lines[i].count(E_letter)
    if countS > 0:
        index=lines[i].index(S_letter)
        Start=(index,i)
        lines[i]=lines[i].replace("S","a")
    if countE > 0:
        index=lines[i].index(E_letter)
        Target=(index,i)
        lines[i]=lines[i].replace("E","z")
   
    
print("S:",Start,"\nE:",Target)
#print(lines)


def Vertice_dir(vertice,inputmap):  #should be able to optimize this function
    i=vertice[0]
    j=vertice[1]
    move={(i-1,j),(i+1,j),(i,j-1),(i,j+1)}
    if i == 0 and j == 0:
        move={(i+1,j),(i,j+1)}
    elif i == 0 and j == len(inputmap)-1:
        move={(i+1,j),(i,j-1)}
    elif i == 0:
        move={(i+1,j),(i,j-1),(i,j+1)}
    elif i == len(inputmap[0])-1 and j == 0:
        move={(i-1,j),(i,j+1)}
    elif i == len(inputmap[0])-1 and j == len(inputmap)-1:
        move={(i-1,j),(i,j-1)}
    elif i == len(inputmap[0])-1:
        move={(i-1,j),(i,j-1),(i,j+1)}
    elif j == 0:
        move={(i-1,j),(i+1,j),(i,j+1)}
    elif j == len(inputmap)-1:
        move={(i-1,j),(i+1,j),(i,j-1)}
    else:
        move={(i-1,j),(i+1,j),(i,j-1),(i,j+1)}
    return move
    
    
def make_Adj(inputmap):  
    Adj={}
    for j in range(len(inputmap)):
        for i in range(len(inputmap[0])):
            move = Vertice_dir((i,j),inputmap)
            if i>162:
                print(i,j,move) ########
            Adj[(i,j)]=[]
            for k in move:
                if abs(ord(inputmap[k[1]][k[0]])-ord(inputmap[j][i]))<=1:
                    Adj[(i,j)].append(k)
                elif ord(inputmap[k[1]][k[0]])<ord(inputmap[j][i]):
                    Adj[(i,j)].append(k)
    return Adj
           
print("length each line:",len(lines[0])) 
Adj = make_Adj(lines)
#print(Adj)

def BFS(S,E,Adj):
    level={S:0}
    parent={S:None} #optional
    i=1
    frontier=[S]
    while frontier:
        next=[]
        step="None"
        for u in frontier:
            for v in Adj[u]:
                #print("level:",i)
                #print(v,E)
                if v not in level:
                    if v == E:
                        step=i
                        return step
                    else:
                        level[v]=i
                        parent[v]=u
                        next.append(v)
        frontier=next
        i+=1
        
    return step

mystep = BFS(Start, Target, Adj)
print(mystep)
```

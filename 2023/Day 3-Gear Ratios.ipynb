```python
# get python re that represents symbols that are not digits and "."
import re
pattern_s = re.compile(r'[^\d.]')
pattern_d = re.compile(r'\d+')

# get the number of rows and columns
rows = len(lines)
columns = len(lines[0].strip())

# # create a 2D list to store the map
map = []
for i in range(rows):
    map.append([0]*columns)

# # fill the map with the input data
# for i in range(rows):
#     for j in range(columns):
#         map[i][j] = lines[i][j]

# #print map in 2D
# for i in range(rows):
#     print(map[i])

def find_symbol(line):
    index = []
    m = re.finditer(pattern_s, line)
    for match in m:
        index.append(match.start())
    return index

def find_digit(line):
    index = []
    m = re.finditer(pattern_d, line)
    for match in m:
        index.append((match.start(), match.end()))
    return index

i = 0
s_list = []
for line in lines:
    line = line.strip()
    symbol = find_symbol(line)
    for j in symbol:
        s_list.append((i, j))
    i += 1

print(s_list)
```

```python
#create list of coordinates that are adjacent to the symbol

def find_adjacent(s_list):
    adjacent = []
    for s in s_list:
        i = s[0]
        j = s[1]
        if i > 0:
            adjacent.append((i-1, j))
        if i < rows-1:
            adjacent.append((i+1, j))
        if j > 0:
            adjacent.append((i, j-1))
        if j < columns-1:
            adjacent.append((i, j+1))
        if i > 0 and j > 0:
            adjacent.append((i-1, j-1))
        if i < rows-1 and j < columns-1:
            adjacent.append((i+1, j+1))
        if i > 0 and j < columns-1:
            adjacent.append((i-1, j+1))
        if i < rows-1 and j > 0:
            adjacent.append((i+1, j-1))
    #remove duplicates
    adjacent = list(set(adjacent))
    return adjacent   

adjacent = find_adjacent(s_list)
print(adjacent)
```

```python
# make a dictionary of digits and their coordinates

d_dict = {}

def get_dict(lines):
    i = 0
    for line in lines:
        m = re.finditer(pattern_d, line)
        for match in m:
            d_dict[ (i, (match.start(), match.end())) ] = match.group()
        i += 1
    return d_dict

d_dict = get_dict(lines)

# print(d_dict)
# get the digit coordinates
# check if the digit coordinates fall within the adjacent coordinates

d_list = []
i = 0
for line in lines:
    digit = find_digit(line)
    for d in digit:
        d_list.append((i, d))
    i += 1

# print(d_list)
# print(len(d_list))

# check if the digit coordinates fall within the adjacent coordinates
adj_digit = []

for d in d_list:
    #get the range of d[1]
    for i in range(d[1][0], d[1][1]):
        if (d[0], i) in adjacent:
            # print(d[0], i)
            # print(d)
            # print((i, (d[1][0], d[1][1])))
            #
            adj_digit.append(int(d_dict[d]))
            break
        i+=1


# print(adj_digit)
print(sum(adj_digit))
```

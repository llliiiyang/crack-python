```python
myfile=open("/Users/yang/Desktop/adventofcode/treematrix.txt",'r')
lines=myfile.readlines()
treelist=[]
#column_list=[]

for y in range(len(lines)):
    treelist.append("")
    treelist[y]=[]
    for x in range(len(lines[y].strip())):
        this_key=lines[y][x]
        this_dict={this_key:4}
        treelist[y].append(this_dict)

#print(treelist)

hidden=0

for y in range(len(treelist)):
    x=0
    v_max=0
    while x < len(treelist[0])-1:
        trh=int(list(treelist[y][x])[0])
        if trh>v_max:
            v_max=trh
        else:
            treelist[y][x].update((j,k-1) for j,k in treelist[y][x].items())
        #print(y,x,treelist[y][x])
        x+=1
    
    x=len(treelist[0])-1
    v_max=0
    while x>0:
        trh=int(list(treelist[y][x])[0])   
        if trh>v_max:
            v_max=trh
        else:
            treelist[y][x].update((j,k-1) for j,k in treelist[y][x].items())
        #print(y,x,treelist[y][x])
        x-=1
                
for x in range(len(treelist[0])):
    y=0
    v_max=0
    while y < len(treelist)-1:
        trh=int(list(treelist[y][x])[0])
        if trh>v_max:
            v_max=trh
        else:
            treelist[y][x].update((j,k-1) for j,k in treelist[y][x].items())
        #print(x,y,treelist[y][x])
        y+=1
       
    y=len(treelist)-1
    v_max=0
    while y>0:
        trh=int(list(treelist[y][x])[0])
        if trh>v_max:
            v_max=trh
        else:
            treelist[y][x].update((j,k-1) for j,k in treelist[y][x].items())
            if list(treelist[y][x].values())[0] ==0:
                hidden+=1
        #print(x,y,treelist[y][x])
        y-=1

print(hidden)
acc=len(treelist)*len(treelist[0])
visible=acc-hidden
print(visible)
```

# --Part 2--

```python
def treeview_func(x,y,mylist):
    view=1
    add=0
    i=x
    a=mylist[y][x]
    while i>0:
        b=mylist[y][i-1]
        #print("a:",a,"b:",b)
        add+=1
        if a<=b:
            break
        i-=1
    view=view*add
    #print("add:",add,"view:",view)
    
    i=x
    add=0
    while i<len(mylist[0])-1:
        b=mylist[y][i+1]
        #print("a:",a,"b:",b)
        add+=1
        if a<=b:
            break
        i+=1
    view=view*add
    #print("add:",add,"view:",view)
    
    j=y
    add=0
    while j>0:
        b=mylist[j-1][x]
        #print("a:",a,"b:",b)
        add+=1
        if a<=b:
            break
        j-=1
    view=view*add
    #print("add:",add,"view:",view)
    
    j=y
    add=0
    while j<len(mylist)-1:
        b=mylist[j+1][x]
        #print("a:",a,"b:",b)
        add+=1
        if a<=b:
            break
        j+=1
    view=view*add
    #print("add:",add,"view:",view)
    
    return view
```

```python
myfile=open("/Users/yang/Desktop/adventofcode/treematrix.txt",'r')
lines=myfile.readlines()
treelist=[]
#column_list=[]

for y in range(len(lines)):
    treelist.append("")
    treelist[y]=[]
    for x in range(len(lines[y].strip())):
        treeheight=int(lines[y][x])
        treelist[y].append(treeheight)

        
x_loc=""
y_loc=""
max_view=0
for y in range(len(treelist)):
    for x in range(len(treelist[0])):
        view=treeview_func(x,y,treelist)
        if view>max_view:
            max_view=view
            x_loc=x
            y_loc=y

print(x_loc,y_loc,max_view)
```

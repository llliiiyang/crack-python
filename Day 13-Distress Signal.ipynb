```python
import ast

myfile=open("/Users/yang/Desktop/adventofcode/packets_order.txt",'r')
lines=myfile.readlines()

def compare_func(left,right):
    #print(left)
    #print(right)
    if len(left) == 0 and len(right) == 0:
        return "Next"
    elif len(left) ==0:
        return "True"
    elif len(right) == 0:
        return "False"
    else:
        for i in range(len(left)):
            if i >= len(right):
                return "False"
            elif type(left[i]) == int and type(right[i]) == int:
                if left[i] < right[i]:
                    return "True"
                elif left[i] > right[i]:
                    return "False"
                else:
                    continue
            elif type(left[i]) == int and type(right[i]) == list:
                left[i]=[left[i]]
                result=compare_func(left[i],right[i])
                if result == "Next":
                    continue
                else:
                    return result
            elif type(left[i]) == list and type(right[i]) == int:
                right[i]=[right[i]]
                result=compare_func(left[i],right[i])
                if result == "Next":
                    continue
                else:
                    return result
            else:
                result=compare_func(left[i],right[i])
                if result == "Next":
                    continue
                else:
                    return result
                
        if len(right)>len(left):
            return "True"
        else:
            return "Next"
    
i=0
index_list=[]
index=1
acc=0
while i < len(lines):
    pair_l=ast.literal_eval(lines[i].strip())
    pair_r=ast.literal_eval(lines[i+1].strip())
    result = compare_func(pair_l,pair_r)
    #print(result,"\n")
    if result == "True":
        index_list.append(index)
        acc+=index
    i+=3
    index+=1

print(acc)
print(len(index_list))
print(index_list)
    
```

```python
import re

with open('/Users/yang/Desktop/adventofcode/sensor_beacon.txt') as file:
    lines=file.read().splitlines()

regex=r"\d+"

#[[],[]],
"""
def split_list(mylist, size=1):
    length = len(mylist)//size
    arrs = []
    while len(mylist) > size:
        cluster = mylist[:size]
        arrs.append(cluster)
        mylist=mylist[size:]
    arrs.append(mylist)
    return arrs
"""

sensor_list=[]
beacon_list=[]

for line in lines:
    inputlist=list(map(int,re.findall(regex,line)))
    sensor_list.append(inputlist[:2])
    beacon_list.append(inputlist[2:])
    
print("sensor:\n",len(sensor_list),sensor_list,"\nbeacon:\n",beacon_list)


def cal_covered_pos(sensor_list,beacon_list,ty=0):
    covered_pos={}
    for i in range(len(sensor_list)):
        dist=abs(sensor_list[i][0]-beacon_list[i][0])+abs(sensor_list[i][1]-beacon_list[i][1])
        #print(i,dist)
        y_dif=abs(sensor_list[i][1]-ty)
        x_dif=dist-y_dif
        #print(i,x_dif,y_dif)
        if x_dif>0:
            x_min=sensor_list[i][0]-x_dif
            x_max=sensor_list[i][0]+x_dif
            if "xmin" in covered_pos:
                x_min=min(x_min,covered_pos["xmin"])
                covered_pos["xmin"]=x_min
                x_max=max(x_max,covered_pos["xmax"])
                covered_pos["xmax"]=x_max
            else:
                covered_pos["xmin"]=sensor_list[i][0]-x_dif
                covered_pos["xmax"]=sensor_list[i][0]+x_dif
    
    covered_length= covered_pos["xmax"]-covered_pos["xmin"]+1
    
    return covered_length, covered_pos



def del_SBcovered (sensor_list,beacon_list,covered_pos,ty=0):
    SBcovered=[]
    for i in range(len(sensor_list)):
        if sensor_list[i][1]==ty and covered_pos["xmin"]<=sensor_list[i][0]<=covered_pos["xmax"]:
    
            print(sensor_list[i][1])
            SBcovered.append(sensor_list[i][0])
             
        if beacon_list[i][1]==ty and covered_pos["xmin"]<=beacon_list[i][0]<=covered_pos["xmax"]:
            
            print(beacon_list[i][1])
            SBcovered.append(beacon_list[i][0])
    SBcovered_num=len(set(SBcovered))
    return SBcovered_num
 
    
    

covered_length, covered_pos = cal_covered_pos(sensor_list,beacon_list,10)
print("covered_length:",covered_length, "covered_pos:",covered_pos)
SBcovered_num = del_SBcovered(sensor_list,beacon_list,covered_pos, 10)

print("covered:",covered_length-SBcovered_num)
    
```

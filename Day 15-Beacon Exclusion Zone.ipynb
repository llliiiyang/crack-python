```python
import re

with open('/Users/yang/Desktop/adventofcode/sensor_beacon.txt') as file:
    lines=file.read().splitlines()

regex=r"\d+"

#[[],[]],
"""
def split_list(mylist, size=1):
    length = len(mylist)//size
    arrs = []
    while len(mylist) > size:
        cluster = mylist[:size]
        arrs.append(cluster)
        mylist=mylist[size:]
    arrs.append(mylist)
    return arrs
"""

sensor_list=[]
beacon_list=[]

for line in lines:
    inputlist=list(map(int,re.findall(regex,line)))
    sensor_list.append(inputlist[:2])
    beacon_list.append(inputlist[2:])
    
print("sensor:\n",sensor_list,"\nbeacon:\n",beacon_list)


def cal_covered_pos(sensor_list,beacon_list,ty=0):
    covered_pos=[]
    for i in range(len(sensor_list)):
        dist=abs(sensor_list[i][0]-beacon_list[i][0])+abs(sensor_list[i][1]-beacon_list[i][1])
        #print(i,dist)
        y_dif=abs(sensor_list[i][1]-ty)
        x_dif=dist-y_dif
        if x_dif>0:
            for k in range (sensor_list[i][0]-x_dif,sensor_list[i][0]+x_dif+1):
                covered_pos.append(k)
    covered_pos= set(covered_pos)
    return covered_pos

def del_SBcovered (sensor_list,beacon_list,covered_pos,ty):
    for i in range(len(sensor_list)):
        if sensor_list[i][1]==ty:
            if  sensor_list[i][0] in covered_pos:
                covered_pos.remove(sensor_list[i][0])
        if beacon_list[i][1]==ty:
            if  beacon_list[i][0] in covered_pos:
                covered_pos.remove(beacon_list[i][0])
            
    covered_pos_num = len(covered_pos)
    return covered_pos_num, covered_pos
    
    
    

covered_pos = cal_covered_pos(sensor_list,beacon_list,10)
covered_pos_num, covered_pos = del_SBcovered(sensor_list,beacon_list,covered_pos,10)

print("covered:",covered_pos_num, covered_pos)
    
```

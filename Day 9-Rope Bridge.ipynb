```python
myfile=open("/Users/yang/Desktop/adventofcode/knots_motions.txt",'r')
lines=myfile.readlines()


def T_move(dict1={"x":0,"y":0},dict2={"x":0,"y":0}):
    
    if dict1["x"]-1<=dict2["x"]<=dict1["x"]+1 and dict1["y"]-1<=dict2["y"]<=dict1["y"]+1:
        #print(dict1,dict2)
        return dict2
    elif dict1["x"]>dict2["x"] and dict1["y"]==dict2["y"]:
        dict2["x"]+=1
    elif dict1["x"]<dict2["x"] and dict1["y"]==dict2["y"]:
        dict2["x"]-=1  
    elif dict1["x"]==dict2["x"] and dict1["y"]>dict2["y"]:
        dict2["y"]+=1
    elif dict1["x"]==dict2["x"] and dict1["y"]<dict2["y"]:
        dict2["y"]-=1
    elif dict1["x"]>dict2["x"] and dict1["y"]>dict2["y"]:
        dict2["x"]+=1
        dict2["y"]+=1
    elif dict1["x"]>dict2["x"] and dict1["y"]<dict2["y"]:
        dict2["x"]+=1
        dict2["y"]-=1
    elif dict1["x"]<dict2["x"] and dict1["y"]>dict2["y"]:
        dict2["x"]-=1
        dict2["y"]+=1
    elif dict1["x"]<dict2["x"] and dict1["y"]<dict2["y"]:
        dict2["x"]-=1
        dict2["y"]-=1
        
    #print(dict1,dict2)
    return dict2

def visits_func(mydict={"x":0,"y":0},myset=set()):
    a=mydict["x"]
    b=mydict["y"]
    myset.add((a,b))
    return myset
    

H_loc={"x":0, "y":0}
T_loc={"x":0,"y":0}
visit_set=set()

for line in lines:
    move=line.split()
    move[1]=int(move[1])
    if move[0]=="R":
        for i in range(move[1]):
            H_loc["x"]+=1
            T_move(H_loc,T_loc)
            visits_func(T_loc,visit_set)
    elif move[0]=="L":
        for i in range(move[1]):
            H_loc["x"]-=1
            T_move(H_loc,T_loc)
            visits_func(T_loc,visit_set)
    elif move[0]=="U":
        for i in range(move[1]):
            H_loc["y"]+=1
            T_move(H_loc,T_loc)
            visits_func(T_loc,visit_set)
    else:
        for i in range(move[1]):
            H_loc["y"]-=1
            T_move(H_loc,T_loc)
            visits_func(T_loc,visit_set)
    #print(H_loc, T_loc)
    

#print(visit_set)
print(len(visit_set))

```
